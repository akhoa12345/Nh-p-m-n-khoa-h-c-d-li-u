## Making a question? (1.25 points)

After exploring the data, we have a better understanding of the data. Now let's see what questions can be answered with this data.

**A possible question is:** What is the population situation of the countries on the continent today?

**Answering this question will** help us to see the current population status and predictions about the future development of the labor force of the countries on the continent are under consideration.

**How we answer this question**: we will consider the Average Birth Rate and Death Rate for countries across all the year.

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
```

### Analyzing to get the answer

Step 01: construct Average Birth Rate dataframe for each country.

```python
# Create a new dataframe
avg_birthrate_df = pd.DataFrame()

# YOUR CODE HERE

avg_birth_rate_series = df_copy.groupby('Country')['Birth Rate'].mean()
avg_birthrate_df = pd.DataFrame({'Average Birth Rate': avg_birth_rate_series})
avg_birthrate_df.reset_index(inplace=True)

avg_birthrate_df

# raise NotImplementedError()
```

```python
# You can just construct plotting for top countries
# YOUR CODE HERE

avg_birthrate_df.sort_values("Average Birth Rate", inplace=True, ascending=False)


plt.figure(figsize=(20, 10))
sns.set(style="whitegrid")

ax = sns.barplot(data=avg_birthrate_df, y="Country", x="Average Birth Rate")


ax.set(xlabel="Average Birth Rate", ylabel="Country")
ax.set_title("Average Birth Rate by Country")

plt.show()
# raise NotImplementedError()
```

Step 02: construct Average Death Rate dataframe for each country.

```python
# Create a new dataframe
avg_deathrate_df = pd.DataFrame()

# YOUR CODE HERE

avg_death_rate_series = df_copy.groupby('Country')['Death Rate'].mean()
avg_deathrate_df = pd.DataFrame({'Average Death Rate': avg_death_rate_series})
avg_deathrate_df.reset_index(inplace=True)

avg_deathrate_df
# raise NotImplementedError()
```

```python
# You can just construct plotting for top countries
# YOUR CODE HERE

avg_deathrate_df.sort_values("Average Death Rate", inplace=True, ascending=False)


plt.figure(figsize=(20, 10))
sns.set(style="whitegrid")

ax = sns.barplot(data=avg_deathrate_df, y="Country", x="Average Death Rate")


ax.set(xlabel="Average Death Rate", ylabel="Country")
ax.set_title("Average Death Rate by Country")

plt.show()
# raise NotImplementedError()
```
