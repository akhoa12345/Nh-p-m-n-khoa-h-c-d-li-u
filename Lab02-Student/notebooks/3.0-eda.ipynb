Dưới đây là một số quan sát dựa trên số liệu thống kê mô tả :

1. **Count**: Hàng "Count" hiển thị số giá trị non-missing cho mỗi numerical feature. Tất cả các đối tượng đều có cùng số lượng, cho ta biết rằng không có giá trị nào bị thiếu trong các cột này.

2. **Mean**: Hàng "Mean" cung cấp giá trị trung bình cho từng feature. Ví dụ: Tổng dân số trung bình là khoảng 90,2 triệu người, Tỷ lệ sinh trung bình là khoảng 20,7 và Tỷ lệ tử vong trung bình là khoảng 6,3.

4. **Minimum and Maximum**: Các hàng "min" và "max" biểu thị giá trị tối thiểu và tối đa cho từng feature. Ví dụ: Tỷ lệ sinh tối thiểu là khoảng 5,1 và tối đa là khoảng 52,1.

5. **Quartiles (25%, 50%, 75%)**: Các phần tư hiển thị các giá trị mà phần trăm dữ liệu nhất định nằm dưới đó. Những số liệu thống kê này cung cấp cái nhìn toàn diện về xu hướng trung tâm, sự phân tán và mối quan hệ giữa các đặc điểm số trong tập dữ liệu. **How we answer this question**: we will consider the Average Birth Rate and Death Rate for countries across all the year.

### Pre-processing

### Analyzing to get the answer

Step 01: construct Average Birth Rate dataframe for each country.

```python
# Create a new dataframe
avg_birthrate_df = pd.DataFrame()

# YOUR CODE HERE

avg_birth_rate_series = df_copy.groupby('Country')['Birth Rate'].mean()
avg_birthrate_df = pd.DataFrame({'Average Birth Rate': avg_birth_rate_series})
avg_birthrate_df.reset_index(inplace=True)

avg_birthrate_df

# raise NotImplementedError()
```

```python
# You can just construct plotting for top countries
# YOUR CODE HERE

avg_birthrate_df.sort_values("Average Birth Rate", inplace=True, ascending=False)


plt.figure(figsize=(20, 10))
sns.set(style="whitegrid")

ax = sns.barplot(data=avg_birthrate_df, y="Country", x="Average Birth Rate")


ax.set(xlabel="Average Birth Rate", ylabel="Country")
ax.set_title("Average Birth Rate by Country")

plt.show()
# raise NotImplementedError()
```

Step 02: construct Average Death Rate dataframe for each country.

```python
# Create a new dataframe
avg_deathrate_df = pd.DataFrame()

# YOUR CODE HERE

avg_death_rate_series = df_copy.groupby('Country')['Death Rate'].mean()
avg_deathrate_df = pd.DataFrame({'Average Death Rate': avg_death_rate_series})
avg_deathrate_df.reset_index(inplace=True)

avg_deathrate_df
# raise NotImplementedError()
```

```python
# You can just construct plotting for top countries
# YOUR CODE HERE

avg_deathrate_df.sort_values("Average Death Rate", inplace=True, ascending=False)


plt.figure(figsize=(20, 10))
sns.set(style="whitegrid")

ax = sns.barplot(data=avg_deathrate_df, y="Country", x="Average Death Rate")


ax.set(xlabel="Average Death Rate", ylabel="Country")
ax.set_title("Average Death Rate by Country")

plt.show()
# raise NotImplementedError()
```
